```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
covid_dataset = pd.read_csv('/content/covid.csv')
```

```python
covid_dataset.head()
```

```python
covid_dataset['location'].value_counts()
```

```python
covid_dataset.info()
```

```python
covid_dataset['date'] = pd.to_datetime(covid_dataset['date'])
```

```python
sns.set(rc={'figure.figsize':{5,10}})
sns.lineplot(x = 'date',y = 'total_cases',data = covid_dataset)
plt.title('Total Cases')
plt.grid(True)
plt.show()
```

```python
sns.set(rc={'figure.figsize':{5,10}})
sns.lineplot(x = 'date',y = 'new_cases',data = covid_dataset)
plt.title('New cases')
plt.grid(True)
plt.show()
```

```python
sns.set(rc={'figure.figsize':{5,10}})
sns.lineplot(x = 'date',y = 'total_deaths',data = covid_dataset)
plt.title('total_deaths')
plt.grid(True)
plt.show()
```

```python
sns.set(rc={'figure.figsize':{5,10}})
sns.lineplot(x = 'date',y = 'new_deaths',data = covid_dataset)
plt.title('New deaths')
plt.grid(True)
plt.show()
```

```python
sns.set(rc={'figure.figsize':{5,10}})
sns.lineplot(x = 'date',y = 'gdp_per_capita',data = covid_dataset,color = 'green')
plt.title('Capital_Gdp')
plt.grid(True)
plt.show()
```

```python
sns.set(rc={'figure.figsize':{5,10}})
sns.lineplot(x = 'date',y = 'population',data = covid_dataset,legend='auto',color = 'red')
plt.title('population')
plt.grid(True)
plt.show()
```
